## Task2: Create table from data from wikipedia page. 

### Use the Notebook to build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe. 

In order to create table of available data from wikipedia page, we need to scrape it by using beautifulsoup package. 

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

we need to convert above table into pandas dataframe. 

```python
df= pd.read_html(tab)
df_Toronto=df[0]
df_Toronto.head()
```

## Task3: Now this dataframe will need some processing and cleaning. 
1. Those postal codes does not have assigned borough will be droped from list. 
2. Those postal codes have borough but does not have neighbourhood assigned : neighbourhood will have same name as borough.
3. merging two borough or neighbourhood with same postal code.

```python
df_Toronto1 = df_Toronto[df_Toronto.Borough != 'Not assigned']


df_Toronto2 = df_Toronto1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df_Toronto2.reset_index(inplace=True)

df_Toronto2['Neighbourhood'] = np.where(df_Toronto2['Neighbourhood'] == 'Not assigned',df_Toronto2['Borough'], df_Toronto2['Neighbourhood'])

df_Toronto2
```

```python
df_Toronto2.shape
```

# Part - 2.

## Use the csv file to create the following dataframe which contains longitudes and latitudes of given postal codes. 

In order to avoid error and long process by using Geocoder package, I am using given csv file with all postal codes with its longitude and latitude. 

```python
Toronto_coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
Toronto_coordinates.head()
```

As we can see, csv file has all postal codes with its coordinates, however we only need those which match with out already created dataframe (df_Toronto2). Therefore, we need to merge this two such a way that it match perfectly. 

```python
df_Toronto3 = pd.merge(df_Toronto2,Toronto_coordinates,on='Postal Code')
df_Toronto3
```

# Part - 3: 

## Explore and cluster the neighborhoods in Toronto.

I am using "Toronto" word as a clustering base. First I will get new dataframe which contains only those borough which has the word "Toronto" in it's name. 
 

```python
df_Toronto4 = df_Toronto3[df_Toronto3['Borough'].str.contains('Toronto',regex=False)]
df_Toronto4
```

```python
df_Toronto4.shape
```

## Plotting above dataframe into folium map. 

First I will get map of Toronto in which all neighbourhoods are marked according to its coordinates. 

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df_Toronto4['Latitude'],df_Toronto4['Longitude'],df_Toronto4['Borough'],df_Toronto4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

# To perform clustering, Kmeans method will be used to create clustering of neighbourhoods of Toronto.

```python
k=5
toronto_clustering = df_Toronto4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df_Toronto4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df_Toronto4
```

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_Toronto4['Latitude'], df_Toronto4['Longitude'], df_Toronto4['Neighbourhood'], df_Toronto4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
